```python
import pandas as pd
import os
os.chdir("../")
path = %pwd
path = os.path.join(path, "artifacts/")
path

data =  pd.read_csv(path + "data.csv")
```

```python
data.drop(columns=['Wafer'], inplace=True)
columns=data.columns
data_n = data.describe()
col_to_drop=[]
for x in columns:
    # print(data_n[x]['std'])
    if (data_n[x]['std'] == 0): # check if standard deviation is zero
        col_to_drop.append(x)  # prepare the list of columns with standard deviation zero
    print(col_to_drop)
```

```python
data.drop(columns=col_to_drop, inplace=True)
data
```

```python
data = data.rename(columns={'Good/Bad': 'Output'})
X = data.drop(columns=['Output'])
Y = data['Output']
```

```python
from sklearn.impute import KNNImputer
import numpy as np
imputer=KNNImputer(n_neighbors=3, weights='uniform',missing_values=np.nan) # we are using KNN imputer
new_data=imputer.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# from kneed import KneeLocator


wcss=[] # initializing an empty list
for i in range (1,50):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=30) # initializing the KMeans object
    kmeans.fit(new_data) # fitting the data to the KMeans Algorithm
    wcss.append(kmeans.inertia_)

print(wcss)

plt.plot(range(1,50),wcss) # creating the graph between WCSS and the number of clusters
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
